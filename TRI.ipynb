# What Has a Neural Representation Actually Learned? A Theory of Representation Invariants (TRI)

## Release I: Invariant Information Decomposition


# Abstract

Modern representation learning has produced remarkably successful models, yet one fundamental question remains unanswered:

> **Given a trained encoder, what invariant information has it actually learned, and what has it irreversibly discarded?**

Existing evaluation methods address this question only indirectly through downstream accuracy, linear probing, representation similarity, or transfer performance. While these approaches measure the usefulness of a representation for specific tasks, they do not characterize the invariant information encoded by the representation itself.

This document introduces the central idea behind the **Theory of Representation Invariants (TRI)**, a mathematical framework for reasoning about learned representations through the invariant information they preserve.

The central result is that every measurable representation admits a unique maximal **invariant core** that captures all invariant information preserved by the encoder. The remaining information forms a complementary variant component. This decomposition provides a representation-centered view of learned features, independent of downstream tasks or performance metrics.

Rather than evaluating representations solely by what they achieve, TRI proposes studying them by **what they preserve**. This document introduces that perspective and motivates the broader mathematical framework developed in subsequent TRI releases.



# 1. The Problem

Representation learning has become a foundational paradigm in modern artificial intelligence, underpinning advances in computer vision, natural language processing, speech, robotics, and multimodal learning. Modern encoders produce representations that achieve remarkable performance across a wide range of downstream tasks.

Despite this empirical success, our understanding of learned representations remains surprisingly limited.

Current evaluation focuses almost exclusively on external behaviour: downstream accuracy, linear probing, transfer learning, representation similarity, retrieval performance, and related benchmarks. These methods assess what a representation can be used for, but they do not characterize the invariant information contained within the representation itself.

As a result, we still lack a mathematical answer to a fundamental question:

> **Given a trained encoder**
>
> $f:X\to Z$,
>
> **what invariant information has it actually preserved, and what has it irreversibly discarded?**

Answering this question requires moving beyond task-dependent evaluation toward a mathematical theory of representations. This is the central motivation behind the **Theory of Representation Invariants (TRI)**.



# 2. Why Existing Evaluation Is Indirect

Existing methods provide valuable insights into learned representations, but they answer questions that are fundamentally different from the one addressed by TRI.

| Method | Measures | Limitation |
|---------|----------|------------|
| Downstream accuracy | Task performance | Depends on a specific prediction task |
| Linear probing | Recoverability of predefined labels | Evaluates only selected concepts |
| CKA / CCA | Similarity between representations | Does not characterize invariant information |
| Transfer learning | Utility on new tasks | Depends on the target task |
| Retrieval metrics | Instance-level similarity | Does not describe the representation's invariant structure |

These methods evaluate a representation through its **behavior**, how well it performs on tasks or how similar it is to other representations.

TRI instead asks a structural question:

> **Which invariant properties of the observations are actually preserved by the representation itself?**

This distinction is fundamental. A representation may achieve excellent downstream performance while preserving very different invariant information from another representation with similar accuracy. TRI focuses on that invariant content directly, rather than on its downstream consequences.



# 3. The Central Idea

A trained encoder can be viewed mathematically as a measurable representation

$$
f:X\to Z.
$$

The central idea of TRI is that every representation possesses a **canonical invariant component** that captures everything the representation preserves with respect to the underlying transformation structure.

Conceptually, every representation can be understood as consisting of two complementary parts:

```text
                  Observation Space X
                          │
                          ▼
                 Trained Encoder f
                          │
                          ▼
                  Representation f
                          │
                          ▼
              Canonical Decomposition
                   f ≅ (φ_f, ν)
                    ╱         ╲
                   ╱           ╲
                  ▼             ▼
      ┌────────────────┐ ┌────────────────┐
      │ Invariant Core │ │  Variant Part  │
      │      φ_f       │ │       ν        │
      └────────────────┘ └────────────────┘
```

The **Invariant Core** ($\phi_f$) is the maximal invariant component contained within the representation. It captures all invariant information that can be recovered from the representation.

The **Variant Part** ($\nu$) contains the remaining information that depends on individual observations rather than invariant structure.

The central claim of TRI is that this decomposition is **canonical**: it is determined entirely by the representation itself and does not depend on downstream tasks, labels, or evaluation procedures.



# 4. The Central Result

The central result of TRI is the **Invariant Information Decomposition Theorem**.

For every measurable representation

$$
f:X\to Z,
$$

there exists a unique (up to equivalence) invariant representation

$$
\phi_f,
$$

called the **Invariant Core**, such that

1. **Factorization.** The invariant core factors through the representation.

2. **Maximality.** Every invariant that factors through the representation also factors through $\phi_f$.

3. **Decomposition.** The representation admits the canonical decomposition

$$
f\cong(\phi_f,\nu),
$$

where $\nu$ contains the remaining variant information.

In essence,

> **Every representation possesses a unique maximal invariant core.**

The invariant core completely characterizes the invariant information preserved by the representation. Everything outside the invariant core belongs to the variant component and is therefore not guaranteed to remain stable under the underlying transformation structure.



# 5. Why This Matters

TRI shifts representation analysis from **performance** to **structure**.

Instead of asking

> *"How accurate is this representation?"*

TRI asks

> *"Which invariant information does this representation preserve?"*

This shift in perspective enables new forms of analysis that do not depend on downstream tasks or benchmark-specific evaluations.

Potential applications include:

- comparing representations through their invariant content,
- understanding the effects of different self-supervised learning objectives,
- analyzing the invariants induced by augmentation strategies,
- diagnosing representation failures before deployment,
- studying robustness under distribution shift,
- guiding the design of representations for safety-critical applications.

Rather than evaluating representations solely by what they achieve on external tasks, TRI provides a mathematical framework for describing what they fundamentally preserve.


# 6. Example

Different representation learning objectives encourage different forms of invariance.

For example, contrastive learning, masked autoencoders, and self-distillation are trained using different transformation or corruption strategies. The TRI framework predicts that these choices lead to different invariant cores because each training objective encourages the representation to preserve a different set of invariant properties.

As a result, two models may achieve similar benchmark performance while preserving fundamentally different invariant information.

Rather than asking only which model achieves higher accuracy, TRI asks a different question:

> **Which invariant information does each representation preserve?**

This perspective suggests comparing representations through their invariant content instead of relying solely on downstream benchmarks or representation similarity measures.

Developing practical methods for estimating and visualizing invariant cores from trained models is an important direction for future work and one of the primary motivations of the TRI framework.



# 7. Toward an Algebra of Representations

The invariant core is only the first component of the **Theory of Representation Invariants (TRI)**.

The broader TRI framework develops a mathematical language for describing and reasoning about learned representations through the invariant information they preserve.

Subsequent releases introduce the remaining components of this framework, including:

- canonical representations,
- representation factorization,
- invariant lattices,
- kernel duality,
- invariant information decomposition,
- representation comparison,
- universal preservation principles,
- theoretical limits of invariant representations.

Together, these components aim to establish a unified algebraic framework for describing, comparing, and understanding learned representations through their invariant structure.
